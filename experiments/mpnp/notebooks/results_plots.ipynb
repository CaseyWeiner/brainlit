```python
import matplotlib.pyplot as plt
import numpy as np 
data_frechet = [[6.5,4.3,3.6,3.3,3.5,3.9,5.8,6.4, 6.7, 5.2, 8.7], [5.2,9.9,3.1,3.0]]
data_sd = [[1.6,2.6,3.3,1.6,2.5,1.9,2.3,2.2, 2.2, 2.1, 2.1], [1.7,2.0,2.0,1.5]]

fig, axes = plt.subplots(1,2)
axes[0].set_title('Frechet Distance', fontsize=14)
axes[0].set_ylabel("Frechet Distance (microns)", fontsize=14)
axes[0].boxplot(data_frechet, labels=["Viterbrain", "APP2"])
axes[0].set_xticklabels(["Viterbrain", "APP2"], fontsize= 14)


axes[1].set_title('Spatial Distance', fontsize=14)
axes[1].set_ylabel("Spatial Distance (microns)", fontsize=14)
axes[1].boxplot(data_sd, labels=["Viterbrain", "APP2"])
axes[1].set_xticklabels(["Viterbrain", "APP2"], fontsize= 14)

plt.sca(axes[0])
plt.yticks(fontsize=14)

plt.sca(axes[1])
plt.yticks(fontsize=14)

fig.set_size_inches(11, 5)
fig.suptitle("Accuracy Metrics of Successful Reconstructions", fontsize=16)
fig.tight_layout()
```
